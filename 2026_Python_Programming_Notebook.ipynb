The continue statement skips the rest of the current turn and jumps straight to the start of the next turn of the loop. It does not stop the loop; it just skips one specific item.

```python
# 1. Build a Tokenizer

import collections

# 1. Helper to find the most common pair of tokens
def get_stats(ids):
    counts = collections.defaultdict(int)
    for pair in zip(ids, ids[1:]):
        counts[pair] += 1
    return counts

# 2. Helper to merge pairs in the sequence
def merge_pair(ids, pair, pair_idx):
    new_ids = []
    i = 0
    while i < len(ids):
        # If we found the pair, merge them into the new token ID
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            new_ids.append(pair_idx)
            i += 2
        else:
            new_ids.append(ids[i])
            i += 1
    return new_ids

# 3. Training the Tokenizer
def train_bpe(text, vocab_size=276):
    tokens = list(text.encode("utf-8"))  # Base vocabulary of 256 bytes
    merges = {}
    
    vocab = {idx: bytes([idx]) for idx in range(256)}
    num_merges = vocab_size - 256

    for i in range(num_merges):
        stats = get_stats(tokens)
        if not stats:
            break
        
        # Get most frequent pair
        best_pair = max(stats, key=stats.get)
        
        # Mint new token ID and add to vocabulary
        new_id = 256 + i
        tokens = merge_pair(tokens, best_pair, new_id)
        
        merges[best_pair] = new_id
        vocab[new_id] = vocab[best_pair[0]] + vocab[best_pair[1]]
        
    return vocab, merges
```

```python
# 4. Cosine Similarity for word Embeddings

import math

# 1. Formula: (A · B) / (||A|| * ||B||)
def cosine_similarity(vec_a, vec_b):
    if len(vec_a) != len(vec_b):
        raise ValueError("Vectors must be of the same dimension.")
        
    dot_product = sum(a * b for a, b in zip(vec_a, vec_b))
    magnitude_a = math.sqrt(sum(a * a for a in vec_a))
    magnitude_b = math.sqrt(sum(b * b for b in vec_b))
    
    if magnitude_a == 0 or magnitude_b == 0:
        return 0.0  # Handle division by zero for empty vectors
        
    return dot_product / (magnitude_a * magnitude_b)

# 2. Mock 4-dimensional Word Embeddings
# Dimensions could represent abstract traits like [Living Thing, Royalty, Femininity, Furry]
embeddings = {
    "king":   [0.9, 0.95, 0.05, 0.0],
    "queen":  [0.9, 0.93, 0.90, 0.0],
    "cat":    [0.95, 0.0, 0.30, 0.9],
    "apple":  [0.1, 0.0, 0.20, 0.0]
}

# 3. Calculate and compare
sim_king_queen = cosine_similarity(embeddings["king"], embeddings["queen"])
sim_king_cat = cosine_similarity(embeddings["king"], embeddings["cat"])
sim_king_apple = cosine_similarity(embeddings["king"], embeddings["apple"])

print(f"Similarity (King, Queen): {sim_king_queen:.4f}")  # Close to 1.0 (Highly Similar)
print(f"Similarity (King, Cat):   {sim_king_cat:.4f}")    # Moderate (Both living things)
print(f"Similarity (King, Apple): {sim_king_apple:.4f}")  # Close to 0.0 (Unrelated)


```

```python
# 5. Mini Bigram Language Model(Next-Word predictor)

import random
from collections import defaultdict, Counter

class BigramLanguageModel:
    def __init__(self):
        # Maps a word -> a Counter of words that follow it
        self.transitions = defaultdict(Counter)
        # Unique start token for sentence generation
        self.start_token = "<START>"
        
    def train(self, corpus):
        """Trains the bigram frequencies using a list of text strings."""
        for text in corpus:
            # Simple tokenization: lowercase and split by whitespace
            words = text.lower().split()
            if not words:
                continue
                
            # Sentence structural mapping: <START> -> first word
            self.transitions[self.start_token][words[0]] += 1
            
            # Map consecutive pairs: word_i -> word_i+1
            for w1, w2 in zip(words, words[1:]):
                self.transitions[w1][w2] += 1

    def predict_next_word(self, current_word):
        """Returns the single most probable next word (Deterministic)."""
        current_word = current_word.lower()
        if current_word not in self.transitions:
            return None
        # Get the word with the highest raw count
        return self.transitions[current_word].most_common(1)[0][0]

    def sample_next_word(self, current_word):
        """Samples the next word weighted by its training probability (Stochastic)."""
        current_word = current_word.lower()
        if current_word not in self.transitions:
            return None
            
        next_word_counts = self.transitions[current_word]
        words = list(next_word_counts.keys())
        weights = list(next_word_counts.values())
        
        # Random choice weighted by frequency
        return random.choices(words, weights=weights)[0]

    def generate_sentence(self, max_length=10):
        """Generates a text sequence starting from the initial model state."""
        sentence = []
        current_word = self.start_token
        
        for _ in range(max_length):
            next_word = self.sample_next_word(current_word)
            if next_word is None:
                break
            sentence.append(next_word)
            current_word = next_word
            
        return " ".join(sentence)

# --- Execution & Testing ---

# Sample training corpus
training_data = [
    "The cat chased the mouse",
    "The dog chased the cat",
    "The mouse ran into the wall",
    "A cat slept on the mat"
]

# Initialize and train model
model = BigramLanguageModel()
model.train(training_data)

# Test 1: Next Word Prediction (Highest Probability)
# "chased" is followed by "the" 2 times.
print("Predict next word after 'chased':", model.predict_next_word("chased")) 

# Test 2: Stochastic Generation
# Generates a pseudo-random sentence based strictly on valid bigram structures
print("\nGenerated Sentences:")
for i in range(3):
    print(f" Pass {i+1}: {model.generate_sentence()}")
```
